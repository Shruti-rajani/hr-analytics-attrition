```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("C:/Users/Rajani Shruti/data-analysis/hr-analytics-attrition/hr_data.csv.csv")
print("Data Loaded succesfully!")

Total_employees = len(df)
print("Total Employees at company:", Total_employees)
```

```python
#attration percentage
attration_percentage = df['Attrition'].value_counts()*100
print("=" * 45)
print("KEY BUSINESS INSIGHTS")
print("=" * 45)
print(f"Attiriton rate:{attration_percentage}")
```

```python
# avrage age of employees who leves the company

attration_age = df[df['Attrition'] == 'Yes']['Age'].mean()
print('Avrage age of employees who left:',attration_age)

stay_age = df[df['Attrition']== 'No']['Age'].mean()
print('Avrage age of Employess who stayed:',stay_age)
```

```python
#job roles having most attration
attrition_job = df[df['Attrition'] == 'Yes']['JobRole'].value_counts()
print('Job Roles where employees leaves:\n', attrition_job)

attrition_job_percent = (df.groupby('JobRole')['Attrition'].apply(lambda x: (x == 'Yes').mean() * 100))
print('\nPercentage of attrition by job role:',attrition_job_percent)

print("="*40)
print("Sales represenatative Job Role has most attration: 39.75%")
```

```python
#employees doing overtime and getting paid less is leaving the comapny

left = df[df['Attrition'] == 'Yes']['OverTime'].value_counts(normalize=True) * 100
print("Employees doing overtime:" ,left)
print("="*40)
income_analysis = df.groupby('Attrition')['MonthlyIncome'].mean()
print("Income of employees:", income_analysis)
```

```python
income_analysis = df.groupby('Attrition')['MonthlyIncome'].mean()
print(income_analysis)

plt.figure(figsize = (4,5))
plt.bar(income_analysis.index, income_analysis.values, color = ['coral','Pink'])
for i, v in enumerate(income_analysis.values):
    plt.text(i, v + 100, f'${v:,.0f}', ha='center', fontweight='bold')
plt.title('Monthly Income of employees')
plt.ylabel('Average Monthly Income ($)')
plt.xlabel('Attrition')
plt.tight_layout()
plt.savefig('income_analysi.png')
plt.show()
```

## Key Finding:

1. Young Employees (age:33) are leaving the comapany.
