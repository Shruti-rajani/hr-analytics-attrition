```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("C:/Users/Rajani Shruti/data-analysis/hr-analytics-attrition/hr_data.csv.csv")
print("Data Loaded succesfully!")
df.head(5)
```

```python
department = df.groupby('Department')['Attrition'].count()
department_attrition = df.groupby('Department')['Attrition'].value_counts()

print(department)
print("\n Department wise Attrition:")
print(department_attrition)

attration_percent = (df.groupby('Department')['Attrition'].apply(lambda x: (x == 'Yes').mean() * 100))
print('\nAttrition Percentage:', attration_percent)

plt.figure(figsize = (6,4))
plt.bar(attration_percent.index, attration_percent.values,color=['steelblue', 'coral', 'green'])
plt.xlabel('Department')
plt.ylabel('Attrition Percentage')
plt.title('Deaprtment-wise Attration')
plt.tight_layout()
plt.savefig('department_wise_attration.png')
plt.show()
```

```python
attration_age = df[df['Attrition'] == 'Yes']['Age'].mean()
print('Avrage age of employees who left:',attration_age)

stay_age = df[df['Attrition']== 'No']['Age'].mean()
print('Avrage age of Employess who stayed:',stay_age)
```

```python
attrition_job = df[df['Attrition'] == 'Yes']['JobRole'].value_counts()
print('Job Roles where employees leaves:\n', attrition_job)

attrition_job_percent = (df.groupby('JobRole')['Attrition'].apply(lambda x: (x == 'Yes').mean() * 100))
print('\nPercentage of attrition by job role:',attrition_job_percent)
```
