```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("C:/Users/Rajani Shruti/data-analysis/hr-analytics-attrition/hr_data.csv.csv")
print("Data Loaded succesfully!")
df.columns
```

```python
#employees who were doing overtime

left = df[df['Attrition'] == 'Yes']['OverTime'].value_counts(normalize=True) * 100
print(left)

plt.figure(figsize = (5,4))
plt.bar(left.index, left.values ,  color = ['steelblue', 'green'])
plt.title('Employees who were doing Overtime and left company')
plt.ylabel('Percentage %')
plt.tight_layout()
plt.show()
```

```python
income_analysis = df.groupby('Attrition')['MonthlyIncome'].mean()
print(income_analysis)

plt.figure(figsize = (4,5))
plt.bar(income_analysis.index, income_analysis.values, color = ['coral','Pink'])
for i, v in enumerate(income_analysis.values):
    plt.text(i, v + 100, f'${v:,.0f}', ha='center', fontweight='bold')
plt.title('Monthly Income of employees')
plt.ylabel('Average Monthly Income ($)')
plt.xlabel('Attrition')
plt.tight_layout()
plt.savefig('income_analysi.png')
plt.show()
```

```python
jobsatisfaction = df.groupby('Attrition')['JobSatisfaction'].mean()
print(jobsatisfaction)

plt.figure(figsize = (4,5))
plt.bar(jobsatisfaction.index, jobsatisfaction.values, color = ['coral','Pink'])
plt.title('Average Job Satisfaction — Left vs Stayed')
plt.ylabel('Average Job Satisfaction (1-4 scale)')
plt.xlabel('Attrition Status')
plt.tight_layout()
plt.savefig('job_satisfaction.png')
plt.show()
```
